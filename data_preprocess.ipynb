```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('survey_responses.xlsx')
df.shape
```

```python
dfc = df.copy()

year_dict = {'1st year': 1, '2nd year': 2, '3rd year': 3, '4th year': 4}
hours_dict = {'1–10 h': 1, '10–20 h': 10, '20-30 h': 20, 'More than 30': 30}
dfc.replace({'What is the year you are currently in your university': year_dict}, inplace=True)
dfc.replace({'How many hours are you spending weekly on e-learning platforms online?': hours_dict}, inplace=True)

dfc
```

```python
year_date = dfc[['What is the year you are currently in your university', 'How many hours are you spending weekly on e-learning platforms online?']]
year_date.columns = ['year', 'time']
year_date
```

```python
year_date.to_csv('year_time.csv', index=False)
```

```python
internet_dict = {'Excellent': 'good', 'Good': 'good', 'Fair': 'good', 'Poor': 'bad', 'Very Poor': 'bad'}
elearning_better_dict = {'Agree': 'agree', 'Strongly agree': 'agree', 'Neutral': 'disagree', 'Disagree': 'disagree', 'Strongly disagree': 'disagree'}
dfc.replace({'How stable is your internet connection, which you used to attend online classes?': internet_dict}, inplace=True)
dfc.replace({'Education through e-learning platforms is better than traditional learning.': elearning_better_dict}, inplace=True)
```

```python
internet_elearning = dfc[['How stable is your internet connection, which you used to attend online classes?', 'Education through e-learning platforms is better than traditional learning.']]
internet_elearning.columns = ['internet', 'better_elearning']
internet_elearning
```

```python
internet_elearning.groupby('internet').count()
```
